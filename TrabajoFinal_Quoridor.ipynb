## **Algoritmos**

Principalmente, se necesita un algoritmo del tipo path-finding para el juego que desarrollaremos, ya que es esencial para que podamos encontrar el camino más corto para avanzar por el tablero. Para este feature, tenemos las siguientes alternativas de algoritmos para implementar:

* ### **Algoritmo de Dijkstra**
Construye el camino más corto posible que une a todos los nodos de un grafo, teniendo en cuenta que la longitud entre cada uno de los nodos, sea la ruta más corta posible.

Dijkstra.png

* ### **Algoritmo de Prim**
El algoritmo de Prim es muy parecido a Djkstra al encontrar el camino más corto posible en un grafo donde los nodos tienen pesos, la diferencia es que este algoritmo toma como prioridad reducir a la menor cantidad posible, el peso de los caminos entre los nodos, esto no significa necesariamente que vaya a ser el camino de menor longitud entre cada par de nodos.

Prim.png

* ### **Algoritmo Minimax**
El algoritmo MiniMax, se encarga de establecer los turnos de cada jugador (jugador y bot) en base al estado en el que se encuentre. Luego de determinar el inicio del turno de la IA, tomará la mejor decisión posible según los obstáculos que encuentre en sus validaciones.

Min1.png

Max2.png
* ### **Algoritmo BFS**
El algoritmo de búsqueda en anchura o BFS(breadth-first search), tiene como propósito encontrar el camino más corto, dado un nodo de origen y en relación a los nodos más próximos recientes. El recorrido mantendrá un orden consecutivo, hasta alcanzar finalmente y por única vez al último nodo del árbol. 

BFS.png

* ### **Algoritmo DFS**
El algoritmo de búsqueda en profundidad o DFS(Depth-First Search), tiene como finalidad, hallar la manera más corta de alcanzar todos los nodos pertenecientes a un grafo desde el nodo del origen. Consiste en expandir todos los caminos hasta recorrer finalmente todos y cada uno de los nodos en un camino concreto

DFS.png






## **Sustentación**

Todos los algoritmos anteriormente mencionados pueden servir para que el bot pueda encontrar el camino más corto hacia el otro lado del tablero y de esta manera ganar el juego, pero cada uno tiene sus diferencias y ventajas de usar. El algoritmo de Prim junto con el algoritmo de Dijkstra asignan pesos a las aristas de un grafo y de esta manera determinar el camino más corto de acuerdo al menor peso total recorrido. Por un lado,  Dijkstra es un algoritmo muy preciso para determinar el camino más corto, pero también algo complicado porque además de eso, se tienen que programar un algoritmo para asignar los valores a las aristas y por qué darles ese valor. Entonces, es muy completo pero a su vez necesita de muchas cosas para ser implementado. Y el de Prim no siempre encuentra el camino más corto pero se acerca.

En cuanto al algoritmo Minimax, tiene una complejidad alta en comparación con otros algoritmos de pathfinding, y también es muy completo. Serviría para un bot con AI, para una mayor dificultad a la hora de jugar.

Otro algoritmo a tomar en cuenta, puede ser el DFS, ya que se encarga de buscar todas las soluciones posibles dentro de cada una de las posibilidades, no es óptimo al referirse a una solución simple y de corto camino, pero sí al tener que recorrer por completo un grafo. Lo preocupante de este algoritmo, es su complejidad de orden exponencial.

Finalmente, el algoritmo BFS se acopla muy bien a nuestras necesidades ya que no buscamos que el bot tenga una dificultad alta, solo que cumpla con el objetivo del juego e intente ganar de la mejor manera posible; y justamente eso nos brinda este algoritmo, ya que es sencillo de implementar y de entender, además nos brinda una opción muy cercana al camino más corto, sino el camino más corto. En conclusión, el algoritmo perfecto para programar una versión sencilla de Quoridor y totalmente funcional.

```
from collections import deque
import resource, sys
import networkx
import matplotlib.pyplot as plt
%matplotlib inline

n = 4 
rows = n
columns = n

Map = networkx.Graph()
Map.add_nodes_from((i,j) for i in range(rows) for j in range(columns))
Map.add_edges_from((((i,j),(i - 1, j)) for i in range(rows) for j in range(columns) if i > 0))
Map.add_edges_from((((i,j),(i, j - 1)) for i in range(rows) for j in range(columns) if j > 0))

def BruteForce(inicio, fin):
    cola = deque()
    cola.append(inicio)
    def Brute_Force(inicio, fin):
        if inicio == fin: return

        def invalidar(x,y):
            return (x < 0) or (x >= rows) or (y < 0) or (y >= columns)

        nodoauxiliar = inicio
        dX, dY = 0, 0

        if nodoauxiliar[0] < fin[0]:
            dX+=1
        elif nodoauxiliar[0] > fin[0]:
            dX-=1
        elif nodoauxiliar[0] == fin[0]:
            if nodoauxiliar[1] < fin[1]:
                dY+=1
            if nodoauxiliar[1] > fin[1]:
                dY-=1
        
        newX, newY = nodoauxiliar[0] + dX,   nodoauxiliar[1] + dY

        if(not invalidar(newX,newY)) and (Map.has_edge(nodoauxiliar,(newX,newY))):
            cola.append((newX,newY))
            Brute_Force((newX,newY), fin)

    Brute_Force(inicio,fin)
    return cola


BruteForce((0,0),(2,3))

```

Después de varias horas de intenso trabajo investigativo y experimental llegamos a la conclusiones de los siguientes puntos:



1.   Quoridor es de los mejores juego para experimentar y entender los distintos algoritmos de búsqueda de camino más corto. Esto debido a la libertad al momento de elegir el camino, y la mecánica de bloquear caminos hacen al Quoridor una de las mejores maneras para captar la idea.
2.   Un tablero cuadrado facilita la implementación de grafos con matrices de coordenadas.
3.   En caso no se encuentren las paredes o algún tipo de obstaculo en el camino, la diferencia entre estos algoritmos podría no ser muy notoria.
4.   Hemos concluido que el mejor algoritmo, en general, para el juego Quoridor es: BFS, debido a que busca el camino más ancho y no recorre todo el tablero para hallar el camino más corto como lo hacen el DFS y la Fuerza Bruta.
5.  La puesta de paredes no dificulta mucho la capacidad de los algoritmos de encontrar los caminos más cortos, aunque si presenta aumentos mínimos en el tiempo de ejecución.
